```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = Image.open('a.jpg')
img_cv = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
mean = cv2.blur(img_cv, (5, 5))
mean_img = Image.fromarray(cv2.cvtColor(mean, cv2.COLOR_BGR2RGB))


plt.imshow(mean_img)
plt.axis('off')
plt.show()
```

1.4 chuyển ảnh sang không gian màu HSV và tách riêng kênh Hue, Saturation, Value để lưu thành ba ảnh grayscale tương ứng

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = Image.open('a.jpg')
img_cv = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
hsv = cv2.cvtColor(img_cv, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

for channel, name in zip([h, s, v], ['a_hue.jpg', 'a_saturation.jpg', 'a_value.jpg']):
    im = Image.fromarray(channel)
    im.save(name)
    plt.imshow(im, cmap='gray')
    plt.axis('off')
    plt.title(name)
    plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
filenames = ['image1.jpg', 'image2.jpg', 'image3.jpg']
images = [cv2.imread(f) for f in filenames]

# Inverse
def inverse(img):
    return 255 - img

# Gamma Correction
def gamma_correction(img):
    gamma = random.uniform(0.5, 2.0)
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255 for i in range(256)]).astype("uint8")
    return cv2.LUT(img, table)

# Log Transform
def log_transform(img):
    img_float = img.astype(np.float32)
    c = random.uniform(1.0, 5.0)
    log_img = c * np.log1p(img_float)
    log_img = np.uint8(cv2.normalize(log_img, None, 0, 255, cv2.NORM_MINMAX))
    return log_img

# Histogram Equalization
def hist_eq(img):
    if len(img.shape) == 2:
        return cv2.equalizeHist(img)
    else:
        ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
        ycrcb[:, :, 0] = cv2.equalizeHist(ycrcb[:, :, 0])
        return cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2BGR)

# Contrast Stretching
def contrast_stretch(img):
    min_val = random.randint(0, 100)
    max_val = random.randint(150, 255)
    img = np.clip(img, min_val, max_val)
    stretched = np.uint8(255 * (img - min_val) / (max_val - min_val + 1e-5))
    return stretched

#  Adaptive Hist Equalization 
def adaptive_hist_eq(img):
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    if len(img.shape) == 2:
        return clahe.apply(img)
    else:
        lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
        lab[:, :, 0] = clahe.apply(lab[:, :, 0])
        return cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

# Danh sách 
options = {
    'I': ('inverse', inverse),
    'G': ('gamma', gamma_correction),
    'L': ('log', log_transform),
    'H': ('hist', hist_eq),
    'C': ('contrast', contrast_stretch),
    'A': ('adaptive', adaptive_hist_eq)
}

print("Nhấn phím để chọn biến đổi:")
print("I - Inverse")
print("G - Gamma Correction")
print("L - Log Transformation")
print("H - Histogram Equalization")
print("C - Contrast Stretching")
print("A - Adaptive Histogram Equalization")
print("Q - Thoát")

while True:
    key = input("Chọn thao tác (I/G/L/H/C/A hoặc Q): ").upper()

    if key == 'Q':
        print("Đã thoát.")
        break

    if key not in options:
        print("Không hợp lệ, hãy thử lại.")
        continue

    name, func = options[key]

    for i, img in enumerate(images):
        img_result = func(img)
        fname = f"output_{name}_{i+1}.jpg"
        cv2.imwrite(fname, img_result)
        img_rgb = cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.title(fname)
        plt.axis('off')
        plt.show()
        print(f"Đã lưu: {fname}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Xử lý ảnh fruits: tăng kích thước thêm 30px
img1 = cv2.imread("colorful-ripe-tropical-fruits.jpg")
img1_out = None
if img1 is not None:
    h1, w1 = img1.shape[:2]
    img1_out = cv2.resize(img1, (w1 + 30, h1 + 30))
    cv2.imwrite("result_resize_fruits.jpg", img1_out)
else:
    print("Không tìm thấy ảnh fruits")

# 2. Xử lý ảnh quang-ninh: xoay 45 độ và lật ngang
img2 = cv2.imread("quang-ninh.jpg")
img2_out = None
if img2 is not None:
    h2, w2 = img2.shape[:2]
    M = cv2.getRotationMatrix2D((w2 // 2, h2 // 2), -45, 1.0)
    rotated = cv2.warpAffine(img2, M, (w2, h2))
    img2_out = cv2.flip(rotated, 1)
    cv2.imwrite("result_rotate_flip_quangninh.jpg", img2_out)
else:
    print("Không tìm thấy ảnh quang-ninh")

# 3. Xử lý ảnh pagoda: phóng to 5 lần + blur + chỉnh sáng/contrast
img3 = cv2.imread("pagoda.jpg")
img3_blur = img3_final = None
if img3 is not None:
    img3_up = cv2.resize(img3, None, fx=5.0, fy=5.0)
    img3_blur = cv2.GaussianBlur(img3_up, (7, 7), 0)
    a = 1.8
    b = 30
    img3_final = np.clip(a * img3_blur + b, 0, 255).astype(np.uint8)
    cv2.imwrite("result_blur_pagoda.jpg", img3_blur)
    cv2.imwrite("result_final_pagoda.jpg", img3_final)
else:
    print("Không tìm thấy ảnh pagoda")

# 4. Hiển thị các ảnh đầu ra bằng matplotlib (RGB)
plt.figure(figsize=(15, 5))

if img1_out is not None:
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img1_out, cv2.COLOR_BGR2RGB))
    plt.axis('off')

if img2_out is not None:
    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(img2_out, cv2.COLOR_BGR2RGB))
    plt.axis('off')

if img3_final is not None:
    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(img3_final, cv2.COLOR_BGR2RGB))
    plt.axis('off')

plt.tight_layout()
plt.show()
```
