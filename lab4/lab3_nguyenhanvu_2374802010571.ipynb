1.5 Dilation và Erosion

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt
data = iio.imread('world_cup.jpg', mode='F')
print(data.shape)
d1 = nd.binary_dilation(data)
plt.imshow(d1)
plt.show()
d2 = nd.binary_dilation(data, iterations=3)
plt.imshow(d2)
plt.show()
```

1.6 Coordinate Mapping

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('world_cup.jpg', mode='F')
print(data.shape)
V, H = data.shape
M = np.indices((V, H))
d = 5
q = 2 * d * np.random.rand(*M.shape) - d
mp = (M + q).astype(int)
d1 = nd.map_coordinates(data, mp)
plt.imshow(d1)
plt.show()
```

1.7 biến đổi chung

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
def GeoFun(outcoord):
    a = 10 * np.cos(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.cos(outcoord[1] / 10.0) + outcoord[1]
    return a, b

data = iio.imread('world_cup.jpg', mode='F')

d1 = nd.geometric_transform(data, GeoFun)
plt.imshow(d1)
plt.show()
 
```

bài tập 5

```python
import numpy as np
import imageio.v2 as iio
import scipy.ndimage as nd
import matplotlib.pyplot as plt

print("Menu:")
print("T - Tịnh tiến")
print("X - Xoay")
print("P - Phóng to")
print("H - Thu nhỏ")
print("C - Coordinate Map")

chon = input("Nhập lựa chọn (T/X/P/H/C): ")

if chon in ['T', 'X', 'P', 'H', 'C']:
    print("Chọn ảnh:")
    print("1 - image1.jpg")
    print("2 - image2.jpg")
    print("3 - image3.jpg")
    so = input("Nhập số 1, 2 hoặc 3: ")
    if so == '1':
        data = iio.imread("exercise/colorful-ripe-tropical-fruits.jpg")
    elif so == '2':
        data = iio.imread("exercise/quang_ninh.jpg")
    else:
        data = iio.imread("exercise/ha-long-bay-in-vietnam.jpg")

    if chon == 'T':
        img = nd.shift(data, (50, 50, 0))
    elif chon == 'X':
        img = nd.rotate(data, 45)
    elif chon == 'P':
        img = nd.zoom(data, (1.5, 1.5, 1))
    elif chon == 'H':
        img = nd.zoom(data, (0.5, 0.5, 1))
    elif chon == 'C':
        h, w, _ = data.shape
        x = np.arange(w)
        y = np.arange(h)
        xv, yv = np.meshgrid(x, y)
        map_x = xv + 20 * np.sin(yv / 20)
        map_y = yv + 20 * np.cos(xv / 20)
        map_x = map_x.astype(np.float32)
        map_y = map_y.astype(np.float32)
        img = nd.map_coordinates(data, [map_y, map_x, np.zeros_like(map_x)], order=1)

    plt.imshow(img)
    plt.axis('off')
    plt.show()
else:
    print("Bạn nhập sai lựa chọn.")
```
